### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
dataFrame.fillna(dataFrame.Age.mean() , inplace = True)
data = dataFrame.copy()
min_val = data.Age.min()
max_val = data.Age.max()
max_val

data['min-max scalling'] = (data.Age - min_val)/(max_val-min_val)
data
```

```python
digits = len(str(int(max_val)))
data['decimal_scalling'] = (data['Age'])/(10**digits)
data
```

```python
mean_val = data.Age.mean()
print("Mean : ",mean_val)
strd_val = data.Age.std()
print("Std Dev : ",strd_val)

data['z_score'] = (data['Age']-mean_val)/strd_val
data
```
